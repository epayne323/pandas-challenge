```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
# there are no null values, so we don't have to worry about removing records with missing data
purchase_data.isnull().any()
```

## Player Count

* Display the total number of players


```python
# there are several ways to get the total number of (unique) players, such as:
# total_players = len(purchase_data["SN"].unique())
purchase_data_SN = purchase_data.set_index("SN")

# this works, but I don't know what ~ means here. ANd I can't get drop_duplicates() to work
unique_players = purchase_data_SN[~purchase_data_SN.index.duplicated(keep='first')]
# unique_players = purchase_data_SN.drop_duplicates()
total_players = unique_players.shape[0]
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# there are also several ways to get these
unique_items = purchase_data["Item Name"].unique()
average_price = purchase_data["Price"].mean()
nPurchases = purchase_data.shape[0]
total_revenue = purchase_data["Price"].sum()
```

```python
# passing grouped object summary series as the values in the dictionary of a new pd.DataFrame({}) 
# will automatically set the index of the new data frame to the groups of those objects, 
# (as long as each is grouped the same way) 
summary_purchases = pd.DataFrame({"Number of Unique Items": [len(unique_items)], "Average Price":[average_price], 
                                  "Number of Purchases":[nPurchases], "Total Revenue":[total_revenue]})
summary_purchases["Average Price"] = summary_purchases["Average Price"].map("${:.2f}".format)
summary_purchases["Total Revenue"] = summary_purchases["Total Revenue"].map("${:.2f}".format)
summary_purchases
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# using the unique_players df from above, which had dropped duplicate SNs
unique_gender_grouped = unique_players.groupby("Gender")

Gender_Demographics = pd.DataFrame({"Total Count": unique_gender_grouped.count()["Age"], 
                                    # percentage calculation uses number of unique players
                                    "Percentage of Players": unique_gender_grouped.count()["Age"]*100/total_players})
                                    
Gender_Demographics["Percentage of Players"] = Gender_Demographics["Percentage of Players"].map("{:.2f}%".format)
Gender_Demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# group by gender and get summary data. this general analysis structure will be repeated throughout this
# notebook.
gender_grouped = purchase_data.groupby("Gender")

purchase_analysis_by_gender = pd.DataFrame({"Purchase Count": gender_grouped.count()["Age"],
                                           "Average Purchase Price":gender_grouped.mean()["Price"],
                                           "Total Purchase Value":gender_grouped.sum()["Price"],
                                           "Avg Total Purchase per Person":gender_grouped.sum()["Price"]/
                                           unique_gender_grouped.count()["Age"]}) # using the grouped object for unique players

# for i in range(1,4):
#     purchase_analysis_by_gender = purchase_analysis_by_gender.iloc[:,i].map("${:.2f}".format)

purchase_analysis_by_gender["Average Purchase Price"] = purchase_analysis_by_gender[
    "Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_gender["Total Purchase Value"] = purchase_analysis_by_gender["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_by_gender["Avg Total Purchase per Person"] = purchase_analysis_by_gender[
    "Avg Total Purchase per Person"].map("${:.2f}".format)

purchase_analysis_by_gender
```

Trends:
Althrough men make up the vast majority of the player base, their average purchases are lower than for women and other/not-specified.

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9,14,19,24,29,34,39,50]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
unique_players.loc[:,"Age Group"] = pd.cut(unique_players["Age"], age_bins, labels=age_labels)
unique_age_grouped = unique_players.groupby("Age Group")

Age_Demographics = pd.DataFrame({"Count": unique_age_grouped.count()["Age"],
                                "Percentage of Players": unique_age_grouped.count()["Age"]/total_players*100})
Age_Demographics["Percentage of Players"] = Age_Demographics[ "Percentage of Players"].map("{:.2f}%".format)
Age_Demographics
```

Trends:
Modal group of player ages is 20-24, followed by 15-19, and 25-29. Together these three groups comprise over 75% of the playerbase.

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# similar analysis as previous section, but purchase-wise instead of player-wise

purchase_data.loc[:, "Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
purchase_age_grouped = purchase_data.groupby("Age Group")

purchase_analysis_by_age = pd.DataFrame({"Purchase Count": purchase_age_grouped.count()["Age"],
                                           "Average Purchase Price":purchase_age_grouped.mean()["Price"],
                                           "Total Purchase Value":purchase_age_grouped.sum()["Price"],
                                           "Avg Total Purchase per Person":purchase_age_grouped.sum()["Price"]/
                                           unique_age_grouped.count()["Age"]}) # using the grouped object for unique players here
purchase_analysis_by_age["Average Purchase Price"] = purchase_analysis_by_age["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_age["Total Purchase Value"] = purchase_analysis_by_age["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_by_age["Avg Total Purchase per Person"] = purchase_analysis_by_age["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis_by_age
```

Trends:
Not really a clear pattern here. Highest-spending group is 35-39, then <10, then 20-24. The oldest group is the lowest-spending. Might be an argument to "hook 'em when they're young". 

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# again, the structure of this section's analysis is similiar to those above
purchase_SN_grouped = purchase_data.groupby("SN")
purchase_analysis_by_SN = pd.DataFrame({"Purchase Count": purchase_SN_grouped.count()["Age"],
                                        "Average Purchase Price":purchase_SN_grouped.mean()["Price"],
                                        "Total Purchase Value":purchase_SN_grouped.sum()["Price"]})
purchase_analysis_by_SN.sort_values("Total Purchase Value", ascending=False, inplace=True)
purchase_analysis_by_SN["Average Purchase Price"] = purchase_analysis_by_SN["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_SN["Total Purchase Value"] = purchase_analysis_by_SN["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_by_SN.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_item_grouped = purchase_data.groupby(["Item ID", "Item Name"])

purchase_analysis_by_item = pd.DataFrame({"Purchase Count":purchase_item_grouped.count()["Purchase ID"],
                                          "Item Price":purchase_item_grouped.first()["Price"], # how
                                          "Total Purchase Value":purchase_item_grouped.sum()["Price"]})
purchase_analysis_by_item.sort_values("Purchase Count", ascending=False, inplace=True)
# saving this before ruining sorting by formatting
top_purchases = purchase_analysis_by_item.sort_values("Total Purchase Value", ascending=False)
purchase_analysis_by_item["Item Price"] = purchase_analysis_by_item["Item Price"].map("${:.2f}".format)
purchase_analysis_by_item["Total Purchase Value"] = purchase_analysis_by_item["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_by_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
top_purchases["Item Price"] = top_purchases["Item Price"].map("${:.2f}".format)
top_purchases["Total Purchase Value"] = top_purchases["Total Purchase Value"].map("${:.2f}".format)
top_purchases.head(20)
```
